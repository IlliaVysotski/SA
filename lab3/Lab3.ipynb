```r
setwd("C:/Users/Illia/Anaconda3/Deep Learning/lab3/")

#install.packages('imager', 'C:/Users/Illia/Anaconda3/envs/r_env/Lib/R/library', repo='https://ftp.uni-sofia.bg/CRAN/')
library(imager) 
library(nortest)

jpegValueScale = 255
histogramStep = 10

img1_path <- "cat_1.jpg"
img2_path <- "cat_2.jpg"

customShow <- function(img, title){
  plot(img, main = title)
}

img_convertion <- function(image_num, img_path) {
  # Original image/grayscale
  img_color <- load.image(img_path)
  customShow(img_color, sprintf("Image %d: color", image_num))
  img_gray <- grayscale(img_color)
  customShow(img_gray, sprintf("Image %d: grayscale", image_num))
  
  # Histogram
  img_gray_scaled <- img_gray * jpegValueScale
  histogram_res <- 
    hist(img_gray_scaled, main=sprintf("Hist of grayscale image %d", image_num), 
         xlab = "", xlim = c(0, jpegValueScale),
         breaks = seq(0, jpegValueScale + histogramStep, by=histogramStep))
  
  list(gray = img_gray_scaled, hist = histogram_res)
}
```

```r
main_calc <- function(image_num, histogram_res) {
  hist_y <- histogram_res$counts
  
  message("Hist ", image_num, " mean = ", mean(hist_y))
  message("Hist ", image_num, " std = ", sd(hist_y))
  message("Hist ", image_num, " median = ", median(hist_y))
  hist_y.t <- table(hist_y)
  message("Hist ", image_num, " mode = ", sort(unique(hist_y))[which.max(hist_y.t)])
}

conv1 <- img_convertion(1, img1_path)
main_calc(1, conv1$hist)
conv2 <- img_convertion(2, img2_path)
main_calc(2, conv2$hist)
```

```r
message("Corrcoef of hist = ", cor(conv1$hist$counts, conv2$hist$counts))
message("Corrcoef of grayscale images = ", cor(conv1$gray, conv2$gray))
```

```r
im1 <- load.image(img1_path)
im2 <- load.image(img2_path)

im_gr1 <- grayscale(im1)
im_gr2 <- grayscale(im2)
im1_range <- jpegValueScale * im_gr1[]
im2_range <- jpegValueScale * im_gr2[]

pearson.test(as.vector(im1_range))
message("Null hypothesis for the first image rejected")
#pearson.test(rnorm(100, mean = 5, sd = 3))
#pearson.test(runif(100, min = 2, max = 4))
pearson.test(as.vector(im2_range))
message("Null hypothesis for the second image rejected")
```
